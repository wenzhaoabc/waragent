```python
import os
from openai import OpenAI

client = OpenAI(
    api_key=os.environ["OPENAI_API_KEY"],
    base_url=os.environ["OPENAI_BASE_URL"],
)


def get_prompt(country):
    return f"""I am conducting a historical simulation study of World War II and need to collect information about the military situation of the countries involved. Here are some factors I need to consider. Please help me write about the situation in {country} at the beginning of the war, specifically in 1939. Please be as detailed as possible, including specific numbers, etc. There may be other military elements that I have not listed, but please include them. Thank you!

Population and Human Resources:
Population structure, including labor force, military recruitment sources, and age distribution.
Conscription system, voluntary military service, and militia organizations.
The impact of the war on population movement and human resources.

Science and Technology Research and Development:
The pace of research and development of weapon systems.
Innovations in military technology, such as radar and rocket technology.
The application of communication and intelligence technology in warfare.

Equipment of Land, Naval, and Air Forces:
Quantities and types of tanks, armored vehicles, artillery, machine guns, pistols, etc.
Numbers, types, and performance of aircraft, including fighters, bombers, and reconnaissance planes.
Naval vessels, including battleships, aircraft carriers, cruisers, destroyers, and submarines.
The production capacity, attrition rates, and replacement speed of various weapon systems.

Military Industry Production:
Output of basic industries such as steel, oil, and coal.
Annual production of firearms, ammunition, artillery, and other military equipment.
The distribution, scale, and annual output of military factories."""


countries = [
    "Germany",
    "Japan",
    "Italy",
    "Hungary",
    "China",
    "Soviet Union",
    "United States",
    "Britain",
    "France",
    "Poland",
]


for country in countries:
    prompt = get_prompt(country)
    response = client.chat.completions.create(
        model="qwen-long",
        messages=[{"role": "user", "content": prompt}],
        stream=False,
    )
    print(response.choices[0].message.content)
    with open(f"temp.py", "a") as f:
        f.write(f'{country.replace(" ","_")} = """{response.choices[0].message.content}""" ')
        f.write("\n\n")
        f.write("===========")
        f.write("\n\n")
```

```python
import os
from openai import OpenAI

client = OpenAI(
    api_key=os.environ["OPENAI_API_KEY"],
    base_url=os.environ["OPENAI_BASE_URL"],
)


def get_prompt(country):
    return f"""I am conducting a historical simulation research on the Second World War and need to collect the financial situation of the participating countries. The following are some factors I need to consider. Please help me write down the situation of {country} in these aspects in the early stage of the war in 1939, in as much detail as possible, including specific values, and please think step by step and give correct and reasonable results, thank you!
1. GDP (Gross Domestic Product): Reflects the overall scale of a nation's economy, encompassing the market value of all final goods and services.
2. Government Revenue:
   - Taxes: Various levies collected from individuals and businesses, forming the main source of government funds.
   - Issuance of National Debt: Raising war funds through the sale of government bonds.
   - Foreign Aid and Loans: Economic support from allies or international organizations.
3. Military Expenditure: Costs directly associated with military maintenance, equipment procurement, operational activities, and military research and development.
4. Social Welfare and Reconstruction Spending:
   - Social Welfare: Subsidies for civilians, pensions for disabled veterans, etc.
   - Post-War Reconstruction: Investments in infrastructure repair and city rebuilding.
5. National and External Debt Levels: The cumulative total of government debt, including both domestic and foreign borrowing.
6. Inflation Rate and Money Supply: The former measures the pace of price increases, while the latter reflects the total amount of currency in circulation.
7. Foreign Exchange Reserves and Exchange Rates: The volume of foreign exchange reserves indicates a country's capacity to repay foreign debts and manage international payments; exchange rate fluctuations affect international trade and payment capabilities.
8. International Trade:
   - Export and Import Volumes: Reflect the country's participation in the global economy and resource dependencies.
   - Strategic Imports of Key Commodities: The situation regarding strategic imports such as oil, food, steel, etc.
9. Gold Reserves: A crucial component of the international monetary system and financial safety net.
10. Productivity and Industrial Output: Particularly the production capability of military supplies, vital to war potential.
11. Human Capital: The direct impact on the economy from the size, skill composition of the workforce, and military mobilization.
12. Tax Policies and Structures: Tax rates, types, and changes affecting government revenue and economic incentives.
13. War Reparations and Compensation: Financial obligations of defeated nations, influencing their fiscal conditions and post-war development.
14. Resource Self-Sufficiency Rate: A country's ability to meet domestic demands internally, affecting reliance on external resources.
15. Agricultural Output: Production of food and key agricultural products, critical for wartime food security.
16. Informal Economy and Black Market Activities: Transactions outside the formal economic system, with complex effects on government revenue and price stability.
17. International Cooperation and Assistance: Economic and military support from international alliances and organizations.
18. War Bonds and Civilian Financing: Public support for the war through bond purchases and other means.
19. Infrastructure Damage and Rehabilitation: Costs of property loss during the war and subsequent reconstruction.
20. Technology Innovation and Military R&D Investment: Costs of new technology development and its potential impact on warfare and the economy.
21. Population Displacement and Refugee Management: The impact of war-induced population movements on the economies and social services of host countries.
22. Wartime Economic Controls: Government policies on resource allocation and price controls, affecting the efficiency of the economy.
"""


countries = [
    "Germany",
    "Japan",
    "Italy",
    "Hungary",
    "China",
    "Soviet Union",
    "United States",
    "Britain",
    "France",
    "Poland",
]


for country in countries:
    prompt = get_prompt(country)
    response = client.chat.completions.create(
        model="qwen-long",
        messages=[{"role": "user", "content": prompt}],
        stream=False,
    )
    print(response.choices[0].message.content)
    with open(f"temp.py", "a") as f:
        f.write(f'{country.replace(" ","_")} = """{response.choices[0].message.content}""" ')
        f.write("\n\n")
        f.write("===========")
        f.write("\n\n")
```

```python
def simplify(text):
    t_f = {
        "GE": "G",
        "JA": "J",
        "IT": "I",
        "HU": "H",
        "CH": "C",
        "SU": "S",
        "US": "U",
        "BR": "B",
        "FR": "F",
        "PO": "P",
    }
    for k, v in t_f.items():
        text = text.replace(k, v)
    text = text.replace("  ", " ")
    return text

text = """
  GE JA IT HU CH SU US BR FR PO
GE /  -  -  -  -  -  -  -  -  -
JA -  /  -  -  -  -  -  -  -  -
IT -  -  /  -  -  -  -  -  -  -
HU -  -  -  /  -  -  -  -  -  -
CH -  -  -  -  /  -  -  -  -  -
SU -  -  -  -  -  /  -  -  -  -
US -  -  -  -  -  -  /  -  -  -
BR -  -  -  -  -  -  -  /  -  -
FR -  -  -  -  -  -  -  -  /  -
PO -  -  -  -  -  -  -  -  -  /
"""

print(simplify(text))
```
